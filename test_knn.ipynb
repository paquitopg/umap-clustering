```python
import knn
import numpy as np 
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
```

```python
X = np.random.multivariate_normal(np.zeros(2), np.identity(2), size=100)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
distance_matrix = squareform(pdist(X, metric="euclidean"))
sigma = -distance_matrix
```

```python
indices, distances = knn.NNDescent(X, sigma, 10, 1000)
```

```python
i = np.random.randint(100)
plt.scatter(X[:, 0], X[:, 1], alpha=.5)
plt.scatter(X[i, 0], X[i, 1])
plt.scatter(X[indices[i], 0], X[indices[i], 1])
plt.show()
```
